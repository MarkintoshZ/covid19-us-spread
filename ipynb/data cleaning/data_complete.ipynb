```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
cts = pd.read_csv('../../data/cleaned/counties.csv')
cts.head()
```

## Map of all counties

```python
fig = go.Figure(data=go.Scattergeo(
        lon = cts['Longitude'],
        lat = cts['Latitude'],
        text = cts['County'] + ', Population: ' + cts['Population'].astype(str),
        mode = 'markers',
        marker_color = np.log(cts.Population),
        ))

fig.update_layout(
        title = 'Counties in US',
        geo_scope='usa',
    )
```

```python
cmt = pd.read_csv('../../data/cleaned/commute_flow.csv')
cmt.head()
```

```python
df = pd.merge(cts, cmt, how='inner', left_on=['State', 'County'], right_on=['Resident State', 'Resident County'])
df.drop(columns=['Resident State', 'Resident County'], inplace=True)
del cts, cmt
df.head()
```
